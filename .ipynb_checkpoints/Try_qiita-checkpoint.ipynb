```python
from gensim import corpora
import csv

with open('p_noun_fkzw.csv','r') as fkzw_nouns:
    reader = csv.reader(fkzw_nouns)

    word_list = []
    for row in reader:
        word_list.append(row)
    
    word_list = word_list
    print(word_list)
```

```python
from gensim import corpora

dictionary = corpora.Dictionary(word_list)
print(dictionary.token2id)
```
