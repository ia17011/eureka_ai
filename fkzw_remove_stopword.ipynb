```python
import csv
import itertools

with open('fukuzawa_just_noun_list.csv','r') as f_just_nouns:
    reader = csv.reader(f_just_nouns)

    word_list = []
    for row in reader:
        word_list.append(row)
    
    word_list = word_list[0]
    print(word_list)
```

```python
with open('slothlib_new.csv', 'r') as slothlib_new:
    stopwords = csv.reader(slothlib_new)
    
    slothlib_list = []
    for stopword in stopwords:
        slothlib_list.append(stopword)
    
for s_word in slothlib_list:
    print(s_word)
```

```python
a = word_list
b = slothlib_list

res = itertools.chain(*b)
res = list(res)

result = [x for x in a if x not in res]

print(result)
with open('p_noun_fkzw.csv','a', newline='') as p_fkzw:
    writer = csv.writer(p_fkzw)
    writer.writerow(result)
```
